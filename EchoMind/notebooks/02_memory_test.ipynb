```python
import sys
sys.path.insert(0, '..')
from src.utils import load_config
config = load_config('../config.yaml')
print('Config loaded:', config['memory'])
```

```python
# Initialize memory
config['memory']['collection_name'] = 'echomind_notebook_test'
config['memory']['persist_directory'] = '../models/saved'

from src.memory import EchoMindMemory
memory = EchoMindMemory(config)
print(f'Memory initialized. Count: {memory.count()}')
```

```python
# Save some test memories
memory.save('The user name is Alex and they love Python')
memory.save('The user is building an AI project called EchoMind')
memory.save('The user prefers dark mode and concise answers')
memory.save('The user wakes up at 6am and exercises every morning')
print(f'Total memories: {memory.count()}')
```

```python
# Test retrieval — semantic search
results = memory.retrieve('What do you know about my project?')
print('Search results:')
for r in results:
    print(f'  → {r}')
```

```python
# Test format for prompt
formatted = memory.format_for_prompt(results)
print('Formatted for prompt injection:')
print(formatted)
```
